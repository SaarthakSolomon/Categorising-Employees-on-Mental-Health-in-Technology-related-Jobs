```python
# Count number of rows
num_rows = len(data)
print(f"Number of rows: {num_rows}")
```

```python
# Define a threshold based on the number of columns (e.g., 50% of the columns)
threshold = len(data.columns) * 0.67

# Drop rows that have more than 50% missing values
data_cleaned = data.dropna(thresh=threshold)
```

```python
# List of columns to apply the change
columns_to_convert = ['Is your employer primarily a tech company/organization?', 'Do you have previous employers?', 'Have you ever sought treatment for a mental health issue from a mental health professional?']

# Use map() to convert 0 -> 'No' and 1 -> 'Yes'
for col in columns_to_convert:
    # Replace 0 with 'No' and 1 with 'Yes'
    data_cleaned[columns_to_convert] = data_cleaned[columns_to_convert].replace({0: 'No', 1: 'Yes'})

# Verify the result
print(data_cleaned.dtypes)
print(data_cleaned)
```

```python
# Get a list of object type columns
categorical_columns = data_cleaned.select_dtypes(include=['object']).columns.tolist()
numeric_columns = data_cleaned.select_dtypes(include=[ 'int64']).columns.tolist()
```

```python
import seaborn as sns
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    return np.sqrt(chi2 / (n * (min(confusion_matrix.shape) - 1)))

# Compute Cramer's V for all categorical features
cramers_v_matrix = pd.DataFrame(index=categorical_columns, columns=categorical_columns)

for col1 in categorical_columns:
    for col2 in categorical_columns:
        cramers_v_matrix.loc[col1, col2] = cramers_v(data[col1], data[col2])

# Convert values to numeric and plot
cramers_v_matrix = cramers_v_matrix.astype(float)
sns.heatmap(cramers_v_matrix, annot=True, cmap="coolwarm")
plt.show()

# Remove highly associated features if needed
```

```python
data_cleaned[categorical_columns]= data_cleaned[categorical_columns].astype(str)
```

```python
# Fill missing values in numerical columns with the mean
data_cleaned[numeric_columns] = data_cleaned[numeric_columns].fillna(data_cleaned[numeric_columns].mean())
```

```python
#fill missing values in categorical columns with the most frequent value(mode)
data_cleaned[categorical_columns]=data_cleaned[categorical_columns].fillna(data_cleaned[categorical_columns].mode().iloc[0])
```

```python
# Check for missing values correctly
print(data_cleaned.isnull().sum())
```

```python
from kmodes.kprototypes import KPrototypes

# K-Prototypes clustering
kproto = KPrototypes(n_clusters=3, init= 'Huang',random_state=42)

#Fit the model and predict clusters (pass categorical column indices)
clusters=kproto.fit_predict(data_cleaned, categorical=[data_cleaned.columns.get_loc(col) for col in categorical_columns])

# Assign the clusters to the Data Frame
data_cleaned['cluster']=clusters
```

```python
# Group by clusters and compute the mean for numerical columns
print(data_cleaned.groupby('cluster')[numeric_columns].mean())

#Group by clusters and find the most frequent value for categorical columns
print(data_cleaned.groupby('cluster')[categorical_columns].agg(lambda x: x.mode().iloc[0]))
```

```python
import matplotlib.pyplot as plt
from kmodes.kprototypes import KPrototypes

costs = []
K_range = range(1, 15)  # Define the range of K (number of clusters)

for k in K_range:
    kproto = KPrototypes(n_clusters=k, init='Huang', random_state=42)
    clusters = kproto.fit_predict(data_cleaned, categorical=[data_cleaned.columns.get_loc(col) for col in categorical_columns])
    costs.append(kproto.cost_)  # K-Prototypes provides a cost value for mixed data

# Plot the elbow curve
plt.plot(K_range, costs, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Cost (Inertia)')
plt.title('Elbow Method for Optimal K (K-Prototypes)')
plt.show()

```
